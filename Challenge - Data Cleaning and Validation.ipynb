```python
clean = welcome['Journal title'].str.replace('J', 'j')
welcome['Journal title clean'] = clean.str.replace('journal', 'j')
welcome['Journal title clean'] = clean.str.replace('plosone', 'plos one')
welcome['Journal title clean'] = clean.str.replace('&', 'and')
```

```python
clean = welcome['Article title'].str.lower()
```

```python
welcome.groupby(clean).count().sort_values(by=['Publisher'], ascending=False)
```

```python
welcome['Article title clean'] = clean.str.replace('\n\nsubjective', 'subjective')
welcome['Article title clean'] = clean.str.replace('\n \n\npersistent', 'persistent')
```

```python
welcome.groupby(welcome['Article title clean']).count().sort_values(by=['Publisher'], ascending=False)
```
